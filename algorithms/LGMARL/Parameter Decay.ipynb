```python
from src.utils.decay import ParameterDecay
import matplotlib.pyplot as plt

comm_eps = ParameterDecay(1.0, 0.01, 100, "sigmoid", 2)

e = [
    comm_eps.get_explo_rate(i)
    for i in range(101)]

plt.figure(figsize=(6, 3))
plt.plot(list(range(0, 10000001, 100000)), e, linewidth=4)
```

```python
import numpy as np
fig = plt.figure()
ax = fig.gca()

for s in [0.5, 1, 1.5, 2.0, 3.0]:
    comm_eps = ParameterDecay(1.0, 0.0, 1e7, "sigmoid", s)

    ids = list(range(0, 10000001, 100000))
    e = [
        comm_eps.get_explo_rate(i)
        for i in ids]

    plt.plot(ids, e, label=r"$\sigma$ = " + str(s), linewidth=2)
    
plt.legend(fontsize=16)
plt.grid()

ax.set_xlabel("Training Step #", fontsize=16)
ax.set_ylabel(r"$\epsilon$ value", fontsize=16)

plt.savefig("/home/doctoq/Desktop/Rdac/LAMAC_figs/comm_eps.svg", bbox_inches="tight")
```
