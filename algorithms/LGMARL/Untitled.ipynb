```python
from src.utils.decay import ParameterDecay
import matplotlib.pyplot as plt

comm_eps = ParameterDecay(1.0, 0.01, 10000, "exp", 2)

e = [
    comm_eps.get_explo_rate(i)
    for i in range(10000)]

plt.plot(e)
```

```python
import numpy as np
fig = plt.figure()
ax = fig.gca()

for s in [0.5, 0.8, 1, 1.2, 1.5, 1.75, 2.0, 2.5, 3.0, 5.0]:
    comm_eps = ParameterDecay(1.0, 0.01, 100, "sigmoid", s)

    e = [
        comm_eps.get_explo_rate(i)
        for i in range(0,101)]

    plt.plot(e, label=str(s))
    
plt.legend()
plt.grid()
ax.set_yticks(np.arange(0, 1.1, 0.1))
ax.set_xticks(np.arange(0, 101, 5))
```

```python
for s in [0.5, 0.8, 1, 2, 3, 5, 8, 10, 15, 20]:
    comm_eps = ParameterDecay(1.0, 0.0, 100, "exp", s)

    e = [
        comm_eps.get_explo_rate(i)
        for i in range(100)]

    plt.plot(e, label=str(s))
plt.legend()
```
