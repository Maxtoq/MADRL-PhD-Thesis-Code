```python
import pandas as pd


lang = pd.read_csv("../../models/magym_PredPrey_RGB/18s50np_lang_ce0/zst_log.csv", index_col=0, usecols=list(range(1, 7)))
noc = pd.read_csv("../../models/magym_PredPrey_RGB/18s50np_noc/zst_log.csv", index_col=0, usecols=list(range(1, 7)))
ec = pd.read_csv("../../models/magym_PredPrey_RGB/18s50np_ec2/zst_log.csv", index_col=0, usecols=list(range(1, 7)))
ae = pd.read_csv("../../models/magym_PredPrey_RGB/18s50np_ec2_ae/zst_log.csv", index_col=0, usecols=list(range(1, 7)))
lg = pd.read_csv("../../models/magym_PredPrey_RGB/18s50np_ec4_lg/zst_log.csv", index_col=0, usecols=list(range(1, 7)))
ae
```

```python
import matplotlib.pyplot as plt

plt.plot(noc["Success rate"], linestyle="--", marker="o")
plt.plot(ec["Success rate"], linestyle="--", marker="o")
plt.plot(ae["Success rate"], linestyle="--", marker="o")
plt.plot(lg["Success rate"], linestyle="--", marker="o")
plt.plot(lang["Success rate"], linestyle="--", marker="o")

plt.ylim(0.85, 1.0)
plt.xticks(list(lang.index), ["AAAA", "AAAB", "AABB", "AABC", "ABCD"])
plt.legend(["No comm", "EC", "EC-AutoEncoder", "EC-LangGround", "LAMARL"], ncols=3)
```
